```python
#dice1 is random dice
#dice2 is designed to generate 2 with greater probability
#dice3 is designed to generate 6 with greater probability
import random
dice=input("enter dice name:")
d={}
count=0
if dice=='dice1':
    print("press enter to roll a dice and q to stop")
    number=input()
    while number!='q':
        num=random.randint(1,6)
        d[num]=d.get(num,0)+1
        print(num)
        number=input()
        count+=1
    l=list(d.keys())
    l.sort()
    for i in l:
        print("probability of "+str(i)+"="+str(round(d[i]/count,2)))
elif dice=='dice2':
    print("press enter to roll a dice and q to stop")
    number=input()
    while number!='q':
        num=random.randint(1,6)
        if num==3:
            num=num-random.randint(0,1)
        elif num==4:
            num=num-random.randint(1,2)
        elif num==5:
            num=num-random.randint(1,3)  
        elif num==6:
            num=num-random.randint(1,4)    
        d[num]=d.get(num,0)+1
        print(num)
        number=input()
        count+=1
    l=list(d.keys())
    l.sort()
    for i in l:
        print("probability of "+str(i)+"="+str(round(d[i]/count,2)))
elif dice=='dice3':
    print("press enter to roll a dice and q to stop")
    number=input()
    while number!='q':
        num=random.randint(1,6)
        if num==1:
            num=num+random.randint(0,1)
        elif num==2:
            num=num+random.randint(0,2)
        elif num==3:
            num=num+random.randint(3,3)
        elif num==4:
            num=num+random.randint(0,2)
        d[num]=d.get(num,0)+1
        print(num)
        number=input()    
        count+=1
    l=list(d.keys())
    l.sort()
    for i in l:
        print("probability of "+str(i)+"="+str(round(d[i]/count,2)))
```
